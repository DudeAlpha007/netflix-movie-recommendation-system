```python
df['tags'] = df['overview'].fillna(' ')+df['keywords'].fillna(' ')+df['genres'].fillna(' ')
df['tags']
```

```python
df['tags'].isna().sum()
```

### Lower case

```python
df['tags'] = df['tags'].apply(lambda x:x.lower())
```

### Punctuations

df['tags'] = df['tags'].str.replace(r'[^\w\s]','',regex = True)
df['tags']

### Stemming

```python
from nltk.stem import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stemming(word):
    y = []
    for i in word.split():
        y.append(ps.stem(i))
    return ' ' .join(y)
```

```python
df['tags'] = df['tags'].apply(stemming)
df['tags']
```

### Converting into vectors

```python
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
```

```python
tf = TfidfVectorizer(max_features=21000,stop_words='english')
```

```python
vector = tf.fit_transform(df['tags'])
```

```python
tf.get_feature_names_out().tolist()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sim = cosine_similarity(vector)
```

```python
sim
```

```python
simdf = pd.DataFrame(sim,index=df['title'],columns=df['title'])
```

```python
simdf
```

```python
simdf['Inception'].sort_values(ascending=False)[1:6]
```

```python
import numpy as np
```

```python
def recommend(movie):
    idx = df[df['title']==movie].index[0]
    movie_list = np.argsort(simdf.iloc[idx])[::-1][1:6]
    return df.iloc[movie_list].title.tolist()
```

```python
recommend('Avatar')
```
